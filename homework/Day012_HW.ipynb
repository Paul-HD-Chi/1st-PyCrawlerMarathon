## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup

import datetime
import re

import pandas as pd

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
# print(f"soup:\n{soup}")
```

### ① 取出今天所有的發文

```python
'''
Your Code
'''
date_pattern = '\d{4}(\/\d{2}){2}'
articles = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    if re.match(date_pattern, d.find(class_="date").text)[0] == datetime.date.today().strftime("%Y/%m/%d"):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        articles.append([d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
columns = ['Datetime', 'Category', 'Title']

df_articles = pd.DataFrame(articles, columns = columns)
df_articles
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
#df_articles.groupby("Category")
cat_articles = pd.DataFrame(df_articles.groupby("Category").size(), columns = ["articles_num"])
cat_articles["rank"] = cat_articles.rank(method = "min", ascending = False)
cat_articles.sort_values(by = "rank")
```
