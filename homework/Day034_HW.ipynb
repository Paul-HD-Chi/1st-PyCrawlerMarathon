```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
proxy_ips = []

'''
Your Code
'''
proxy_ip_url = "https://free-proxy-list.net/"

browser = webdriver.Chrome()
browser.get(proxy_ip_url)  # open browser and connect to the web site
time.sleep(2)  # sleep 2 seconds

while True:
    time.sleep(2)  # sleep 2 seconds
    
    html = browser.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    tr_list = soup.find("tbody").find_all("tr")
    for tr in tr_list:
        # print(f"tr:\n{tr}")
        tds = tr.find_all("td")
        ip = tds[0].text + ":" + tds[1].text
        proxy_ips.append(ip)
        print(f"ip: {ip}")
    
    print("===== next page =====")
    last_page = soup.find('li', attrs = {'class': 'fg-button ui-button ui-state-default next disabled'})
    if last_page:
        print("========== here is the last page ==========")
        break
    else:
        next_page = browser.find_element_by_link_text('Next')
        next_page.click()

print(f"number of ips: {len(proxy_ips)}")
```

```python
# close browser
browser.quit()
```
