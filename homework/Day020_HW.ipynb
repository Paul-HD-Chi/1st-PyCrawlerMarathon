```python
'''
Your Code
'''
d = dict()

trs = table.find_all('tr')
for tr in trs:
    #print(f"tr: {tr}")
    tds = tr.find_all('td')
    #print(f"tds: {tds}")
    
    #print(f"length: {len(tds)}")
    if len(tds) == 5:
        #print(f"tds[0] = {tds[0]}")  # supervised particle
        #print(f"tds[1] = {tds[1]}")  # unit
        #print(f"tds[2] = {tds[2]}")  # month
        #print(f"tds[3] = {tds[3]}\n")  # value
        
        key = tds[0].text
        d[key] = dict()
        d[key][tds[2].text] = tds[3].text
    elif len(tds) == 3:
        #print(f"tds[0] = {tds[0]}")  # month
        #print(f"tds[1] = {tds[1]}")  # valuet
        
        d[key][tds[0].text] = tds[1].text
        
print(d)
```
