### ① 印出最新文章的「作者」「標題」「時間」

```python
'''
Your Code
'''
def DumpData(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html5lib')
    
    for d in soup.find_all(class_="title"):
        print(d.text.replace('\t', '').replace('\n', ''))
        try:
            r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
            meta = r.find_all(class_ = 'article-metaline')
            # print(f"meta:\n{meta}")
            print("作者：", meta[0].find(class_ = 'article-meta-value').text)
            print("標題：", meta[1].find(class_ = 'article-meta-value').text)
            print("時間：", meta[2].find(class_ = 'article-meta-value').text)
            print("="*60)
        except:
            continue
    
url = 'https://www.ptt.cc/bbs/NBA/index.html'
DumpData(url)
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
'''
Your Code
'''
url = 'https://www.ptt.cc/bbs/NBA/index1.html'
DumpData(url)
```
