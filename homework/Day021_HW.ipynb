```python
from selenium import webdriver
from bs4 import BeautifulSoup

import time
from datetime import datetime, timedelta

import pandas as pd
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
''' Paul Chi, for test
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

h3Tags = soup.find(class_="part_list_2").find_all('h3')
# print(f"h3Tags:\n{h3Tags}")

for h3Tag in h3Tags:
    print(f"h3Tag:\n{h3Tag}")
    print(f"h3Tag.span: {h3Tag.span.text}")  # datetime
    print(f"h3Tag.em: {h3Tag.em.text}")  # category
    print(f"h3Tag.a: {h3Tag.a.text}\n")  # subject
'''    
```

## 根據範例，取出今天所有的新聞

```python
'''
Your Code
'''
date = datetime.today().strftime('%Y/%m/%d')

while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    
    h3Tags = soup.find(class_="part_list_2").find_all('h3')
    date_lastnews = h3Tags[-1].span.text
    if date_lastnews[:10] != date:
        break
```

```python
articles = []

for h3Tag in h3Tags:
    if h3Tag.span.text[:10] == date:
        print(h3Tag.span.text, h3Tag.em.text, h3Tag.a.text)
        
        articles.append([h3Tag.span.text, h3Tag.em.text, h3Tag.a.text])
```

```python
columns = ['Datetime', 'Category', 'Title']
df_articles = pd.DataFrame(articles, columns = columns)
df_articles
```

## 取出今天下午三點到五點的新聞

```python
'''
Your Code
'''
upperbound = datetime.strptime("2020/01/01 17:00", "%Y/%m/%d %H:%M")
lowerbound = datetime.strptime("2020/01/01 15:00", "%Y/%m/%d %H:%M")

articlesLimitedPeriod = []

for h3Tag in h3Tags:
    if h3Tag.span.text[:10] == date:
        tagTime = datetime.strptime(h3Tag.span.text, '%Y/%m/%d %H:%M')
        if lowerbound < tagTime < upperbound:
            print(h3Tag.span.text, h3Tag.em.text, h3Tag.a.text)
            
            articlesLimitedPeriod.append([h3Tag.span.text, h3Tag.em.text, h3Tag.a.text])
```

```python
df_articlesLimitedPeriod = pd.DataFrame(articlesLimitedPeriod, columns = columns)
df_articlesLimitedPeriod
```

## 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
now = datetime.now()
threedaysago = (now - timedelta(days = 3)).strftime("%Y-%m-%d")
url = "https://www.ettoday.net/news/news-list-" + threedaysago + "-0.htm"
browser.get(url)
```

```python
date = (now - timedelta(days = 3)).strftime("%Y/%m/%d")

while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    
    h3Tags = soup.find(class_="part_list_2").find_all('h3')
    date_lastnews = h3Tags[-1].span.text
    if date_lastnews[:10] != date:
        break
```

```python
upperbound = datetime.strptime(date + " 17:00", "%Y/%m/%d %H:%M")
lowerbound = datetime.strptime(date + " 15:00", "%Y/%m/%d %H:%M")

articlesThreeDaysAgo = []

for h3Tag in h3Tags:
    if h3Tag.span.text[:10] == date:
        tagTime = datetime.strptime(h3Tag.span.text, '%Y/%m/%d %H:%M')
        if lowerbound < tagTime < upperbound:
            print(h3Tag.span.text, h3Tag.em.text, h3Tag.a.text)
            
            articlesThreeDaysAgo.append([h3Tag.span.text, h3Tag.em.text, h3Tag.a.text])
```

```python
df_articlesThreeDaysAgo = pd.DataFrame(articlesThreeDaysAgo, columns = columns)
df_articlesThreeDaysAgo
```
